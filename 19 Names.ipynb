```python
from nltk.corpus import names
```

```python
names.fileids()
```

```python
len(names.words('female.txt'))
```

```python
len(names.words('male.txt'))
```

```python
#Imprimir nombres por letra
male = names.words('male.txt')
female = names.words('female.txt')
for name in male and female:
    initial = name[0]
    if initial == ('S'):
        print(name)
```
