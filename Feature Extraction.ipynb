```python
doc = nlp('We are going to die, and that makes us the lucky ones. Most people are never going to die because they are never going to be born. The potential people who could have been here in my place but who will in fact never see the light of day outnumber the sand grains of Arabia. Certainly those unborn ghosts include greater poets than Keats, scientists greater than Newton. We know this because the set of possible people allowed by our DNA so massively exceeds the set of actual people. In the teeth of these stupefying odds it is you and I, in our ordinariness, that are here.')
```

###Part-of-Speech [POS] Tagging

```python
for token in doc:
    print(token.text, token.lemma_)
```

```python
#Format columns
for token in doc:
    print(f'{token.text:{15}} {token.lemma_:{15}} {token.pos_:{10}} {token.is_stop}')
```

```python
#Dependency Parsing

for chunk in doc.noun_chunks:
    print(f'{chunk.text:{25}} {chunk.root.text:{15}} {chunk.root.dep_}')
```

```python
#Named Entity Recognition

for ent in doc.ents:
    print(ent.text, ent.label)
```

```python
#Sentence Segmentation

for sent in doc.sents:
    print(sent)
```
